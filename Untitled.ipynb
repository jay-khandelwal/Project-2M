```python
import requests
import random
import time


HEADERS = {
    'Authorization': f'Bearer {API_TOKEN}',
    'Content-Type': 'application/json'
}

# Function to create a shape (node) in Miro
def create_shape(board_id, text, x, y):
    url = f'https://api.miro.com/v2/boards/{board_id}/widgets'
    payload = {
        "data": {
            "content": text
        },
        "style": {
            "shapeType": "rectangle",
            "backgroundColor": "#FFFFFF",
            "borderColor": "#000000"
        },
        "position": {
            "x": x,
            "y": y
        },
        "type": "shape"
    }
    response = requests.post(url, headers=HEADERS, json=payload)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error creating shape: {response.status_code}, {response.text}")
        return None

# Function to create a connector (edge) in Miro
def create_connector(board_id, start_widget_id, end_widget_id):
    url = f'https://api.miro.com/v2/boards/{board_id}/widgets'
    payload = {
        "type": "connector",
        "startWidgetId": start_widget_id,
        "endWidgetId": end_widget_id,
        "style": {
            "lineColor": "#000000",
            "lineStyle": "normal",
            "lineThickness": 1
        }
    }
    response = requests.post(url, headers=HEADERS, json=payload)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error creating connector: {response.status_code}, {response.text}")
        return None

# Create random graph with given number of nodes
def create_random_graph_in_miro(board_id, num_nodes):
    nodes = []
    for i in range(num_nodes):
        x = random.randint(-5000, 5000)
        y = random.randint(-5000, 5000)
        node = create_shape(board_id, f'Node {i}', x, y)
        if node and 'id' in node:
            nodes.append(node)
        else:
            print(f"Failed to create node {i}")
        time.sleep(0.2)  # To avoid rate limits

    edge_density = 0.1  # Adjust density to control number of edges
    for i in range(num_nodes):
        for j in range(i + 1, num_nodes):
            if random.random() < edge_density:
                if 'id' in nodes[i] and 'id' in nodes[j]:
                    create_connector(board_id, nodes[i]['id'], nodes[j]['id'])
                else:
                    print(f"Missing node id for edge between {i} and {j}")

# Generate a random graph with 20 nodes in Miro
num_nodes = 20
create_random_graph_in_miro(BOARD_ID, num_nodes)
```
