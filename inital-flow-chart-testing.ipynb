```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
df = pd.read_json('data.json')
```

```python
filtered_df = df[df["Referer"].notna()]
```

```python
filtered_df.shape
```

```python
for index, row in filtered_df.head(10).iterrows():
    referer = row['Referer']
    target_url = row['path']
    G.add_edge(referer, target_url)
```

```python
# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, iterations=50)
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="skyblue", font_size=10, font_color="black", font_weight="bold", edge_color="gray")
plt.title("Page to API Request Flowchart")
plt.show()
```
