```python
filtered_df = df[df["Referer"].notna()]
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
```

```python
for index, row in filtered_df.iterrows():
    referer = row['Referer']
    target_url = row['path']
    G.add_edge(referer, target_url)

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, iterations=50)
nx.draw(G, pos, with_labels=True, node_size=6000, node_color="gray", font_size=10, font_color="black", font_weight="bold", edge_color="gray")
plt.title("Page to API Request Flowchart")
plt.show()
```

```python
exclude_extensions = ['css', 'js', 'jpeg', 'jpg', 'mp4', 'png']

# Create a regex pattern for the extensions
pattern = '|'.join([f'\\b{ext}\\b' for ext in exclude_extensions])
```

```python
filtered_df = df[df["Referer"].notna() & ~df["file_extension"].str.contains(pattern, case=False, na=False)]
```
